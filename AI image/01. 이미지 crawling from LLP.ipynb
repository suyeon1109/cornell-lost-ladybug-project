```python
## 반복문에 들어갈 변수 (종 명) 뽑아 리스트화하기

import requests
from bs4 import BeautifulSoup

# URL of the web page
webpage_url = "http://www.lostladybug.org/summary-page-555.php"

# Column index (0-based) that you want to extract values from
column_index = 1

# Send a GET request to the web page
response = requests.get(webpage_url)

if response.status_code == 200:
    soup = BeautifulSoup(response.text, "html.parser")

    # Find the table containing the data
    table = soup.find("table")

    if table:
        column_values = []

        # Loop through the rows of the table
        for row in table.find_all("tr"):
            columns = row.find_all("td")
            
            # Check if the row has enough columns
            if column_index < len(columns):
                column_value = columns[column_index].text.strip()
                column_values.append(column_value)

        # Print or manipulate the extracted column values
        print("Column Values:", column_values)
    else:
        print("Table not found on the web page.")
else:
    print(f"Failed to fetch the web page: {webpage_url}")
```

```python
column_values

original_list = column_values
updated_list = []

# Replace elements in the list
for element in original_list:
    words = element.split()

    # Join the first two words with a '+'
    result = '+'.join(words[:2])
    updated_list.append(result)

print("Original List:", original_list)
print("Updated List:", updated_list)
```

```python
import os

# List of folder names
folder_names = updated_list

# Create folders with names from the list
for folder_name in folder_names:
    os.makedirs(folder_name, exist_ok=True)
    print(f"Folder created: {folder_name}")
```

```python
import os
import time
import requests
from bs4 import BeautifulSoup

for element in updated_list:
    image_urls = []

    # URL of the web page
    webpage_url = "http://www.lostladybug.org/contributors-images-species.php?species=" + element

    # CSS selector pattern without the varying number
    css_selector_pattern = "#photos > li:nth-child({}) > div > a > img"

    # Initialize the counter for repetitions
    repetition_count = 0

    # Send a GET request to the web page
    response = requests.get(webpage_url)

    if response.status_code == 200:
        soup = BeautifulSoup(response.text, "html.parser")

        # Fetch all image elements first to avoid multiple requests
        all_image_elements = soup.select("#photos > li > div > a > img")

        # Loop through increasing numbers until the selector doesn't match
        for repetition_count in range(1, len(all_image_elements) + 1):
            current_selector = css_selector_pattern.format(repetition_count)

            # Check if the element exists in the pre-fetched list
            image_element = all_image_elements[repetition_count - 1]

            if image_element:
                image_url = image_element['src']
                print("Image URL:", image_url)
                image_urls.append(image_url)
            else:
                break

            # Introduce a delay to avoid rapid requests
            time.sleep(2)  # Adjust the delay as needed

        # Print the count of repetitions
        print("Number of Repetitions:", len(image_urls))
    else:
        print(f"Failed to fetch the web page: {webpage_url}")

    for image_url in image_urls:
        # Image URL
        if not image_url.startswith("http"):
            image_url = "http://www.lostladybug.org/" + image_url

        # Folder name
        folder_name = element

        # Create the folder if it doesn't exist
        os.makedirs(folder_name, exist_ok=True)

        # Get the image file name from the URL
        image_filename = os.path.basename(image_url)

        # Combine folder path and image filename
        image_path = os.path.join(folder_name, image_filename)

        # Send a GET request to download the image
        response = requests.get(image_url)

        if response.status_code == 200:
            with open(image_path, "wb") as img_file:
                img_file.write(response.content)
            print(f"Image saved to: {image_path}")
        else:
            print(f"Failed to download the image from: {image_url}")
```
